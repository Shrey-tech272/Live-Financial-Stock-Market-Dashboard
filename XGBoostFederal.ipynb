```python
import pandas as pd
import numpy as np
import yfinance as yf
from xgboost import XGBRegressor
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import datetime
import time
import os

# -------------------------------
# Load historical data
# -------------------------------
hist_file = "FEDERALBK_stock_data.csv"
df = pd.read_csv(hist_file)
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
data = df[['Close']].copy()
data.rename(columns={'Close':'close'}, inplace=True)

# -------------------------------
# Technical indicators
# -------------------------------
window = 15

# Returns
data['return'] = data['close'].pct_change()

# Moving averages
data['ma_5'] = data['close'].rolling(5).mean()
data['ma_10'] = data['close'].rolling(10).mean()
data['ma_20'] = data['close'].rolling(20).mean()

# Exponential moving averages
data['ema_5'] = data['close'].ewm(span=5, adjust=False).mean()
data['ema_10'] = data['close'].ewm(span=10, adjust=False).mean()
data['ema_20'] = data['close'].ewm(span=20, adjust=False).mean()

# Volatility
data['vol_5'] = data['close'].rolling(5).std()
data['vol_10'] = data['close'].rolling(10).std()
data['vol_20'] = data['close'].rolling(20).std()

# RSI 14
delta = data['close'].diff()
gain = delta.clip(lower=0)
loss = -delta.clip(upper=0)
avg_gain = gain.rolling(14).mean()
avg_loss = loss.rolling(14).mean()
rs = avg_gain / avg_loss
data['RSI_14'] = 100 - (100 / (1 + rs))

# Lag features
for i in range(1, window+1):
    data[f'lag_{i}'] = data['close'].shift(i)

data.dropna(inplace=True)

feature_cols = [f'lag_{i}' for i in range(1, window+1)] + \
               ['ma_5','ma_10','ma_20','ema_5','ema_10','ema_20',
                'vol_5','vol_10','vol_20','return','RSI_14']

X = data[feature_cols]
y = data['close']

# -------------------------------
# Train XGBoost
# -------------------------------
model = XGBRegressor(
    n_estimators=1000,
    learning_rate=0.03,
    max_depth=4,
    subsample=0.85,
    colsample_bytree=0.85,
    random_state=42
)
model.fit(X, y)

# -------------------------------
# Initialize CSV
# -------------------------------
log_file = 'FEDERALBK_stock_data.csv'
log_columns = ['Date','Live_Price','Next_Day','Next_Week','Next_Month',
               'RMSE_Next_Day','MAPE_Next_Day','RMSE_Next_Week','MAPE_Next_Week',
               'RMSE_Next_Month','MAPE_Next_Month']

if not os.path.exists(log_file):
    pd.DataFrame(columns=log_columns).to_csv(log_file, index=False)

# -------------------------------
# Historical predictions
# -------------------------------
history_prices = list(data['close'])
predictions_log = []

for i in range(window, len(data)):
    current_date = data.index[i]
    current_features = data.iloc[i].copy()
    
    temp_history = history_prices[:i].copy()
    preds = []
    
    for step in [1,5,21]:  # next day, week, month
        input_row = pd.DataFrame({col: [current_features[col]] for col in feature_cols})
        pred_val = model.predict(input_row)[0]
        preds.append(pred_val)
        
        # Update features for next step
        temp_history.append(pred_val)
        for j in range(1, window+1):
            current_features[f'lag_{j}'] = temp_history[-j]
        current_features['return'] = (temp_history[-1] - temp_history[-2])/temp_history[-2]
        current_features['ma_5'] = np.mean(temp_history[-5:])
        current_features['ma_10'] = np.mean(temp_history[-10:])
        current_features['ma_20'] = np.mean(temp_history[-20:])
        current_features['ema_5'] = pd.Series(temp_history[-5:]).ewm(span=5, adjust=False).mean().iloc[-1]
        current_features['ema_10'] = pd.Series(temp_history[-10:]).ewm(span=10, adjust=False).mean().iloc[-1]
        current_features['ema_20'] = pd.Series(temp_history[-20:]).ewm(span=20, adjust=False).mean().iloc[-1]
        current_features['vol_5'] = np.std(temp_history[-5:])
        current_features['vol_10'] = np.std(temp_history[-10:])
        current_features['vol_20'] = np.std(temp_history[-20:])
        delta_temp = np.diff(temp_history[-15:])
        gain = np.clip(delta_temp,0,None)
        loss = -np.clip(delta_temp,None,0)
        avg_gain = np.mean(gain)
        avg_loss = np.mean(loss)
        rs = avg_gain/avg_loss if avg_loss!=0 else 0
        current_features['RSI_14'] = 100 - (100/(1+rs)) if avg_loss!=0 else 100
    
    next_day_price, next_week_price, next_month_price = preds
    
    # Calculate RMSE & MAPE if actual available
    errors = {}
    horizons = ['Next_Day','Next_Week','Next_Month']
    days_ahead = [1,7,21]
    
    for h,d in zip(horizons, days_ahead):
        target_index = i + d
        if target_index < len(history_prices):
            actual = history_prices[target_index]
            pred = preds[horizons.index(h)]
            rmse = np.sqrt((pred - actual)**2)
            mape = abs((pred - actual)/actual)*100
            errors[f'RMSE_{h}'] = rmse
            errors[f'MAPE_{h}'] = mape
        else:
            errors[f'RMSE_{h}'] = np.nan
            errors[f'MAPE_{h}'] = np.nan
    
    row = {
        'Date': current_date,
        'Live_Price': history_prices[i],
        'Next_Day': next_day_price,
        'Next_Week': next_week_price,
        'Next_Month': next_month_price,
        **errors
    }
    predictions_log.append(row)

# Save historical predictions
pd.DataFrame(predictions_log).to_csv(log_file, index=False)
print("✅ Historical predictions Federal.")

# -------------------------------
# Live prediction loop
# -------------------------------
plt.ion()
history_prices_live = list(data['close'])
ticker = "FEDERALBNK.NS"

while True:
    now = datetime.datetime.now()
    
    # Fetch live price
    stock = yf.Ticker(ticker)
    live_data = stock.history(period="1d", interval="1m")
    
    last_price = live_data['Close'][-1] if not live_data.empty else history_prices_live[-1]
    history_prices_live.append(last_price)
    
    # Multi-step prediction
    if len(history_prices_live) >= window:
        temp_history = history_prices_live.copy()
        preds = []
        current_features = data.iloc[-1].copy()
        
        for step in [1,5,21]:
            input_row = pd.DataFrame({col: [current_features[col]] for col in feature_cols})
            pred_val = model.predict(input_row)[0]
            preds.append(pred_val)
            
            temp_history.append(pred_val)
            # update features same as above
            for j in range(1, window+1):
                current_features[f'lag_{j}'] = temp_history[-j]
            current_features['return'] = (temp_history[-1] - temp_history[-2])/temp_history[-2]
            current_features['ma_5'] = np.mean(temp_history[-5:])
            current_features['ma_10'] = np.mean(temp_history[-10:])
            current_features['ma_20'] = np.mean(temp_history[-20:])
            current_features['ema_5'] = pd.Series(temp_history[-5:]).ewm(span=5, adjust=False).mean().iloc[-1]
            current_features['ema_10'] = pd.Series(temp_history[-10:]).ewm(span=10, adjust=False).mean().iloc[-1]
            current_features['ema_20'] = pd.Series(temp_history[-20:]).ewm(span=20, adjust=False).mean().iloc[-1]
            current_features['vol_5'] = np.std(temp_history[-5:])
            current_features['vol_10'] = np.std(temp_history[-10:])
            current_features['vol_20'] = np.std(temp_history[-20:])
            delta_temp = np.diff(temp_history[-15:])
            gain = np.clip(delta_temp,0,None)
            loss = -np.clip(delta_temp,None,0)
            avg_gain = np.mean(gain)
            avg_loss = np.mean(loss)
            rs = avg_gain/avg_loss if avg_loss!=0 else 0
            current_features['RSI_14'] = 100 - (100/(1+rs)) if avg_loss!=0 else 100
        
        next_day_price, next_week_price, next_month_price = preds
        
        # Save live prediction
        new_row = {
            'Date': now,
            'Live_Price': last_price,
            'Next_Day': next_day_price,
            'Next_Week': next_week_price,
            'Next_Month': next_month_price
        }
        pd.DataFrame([new_row]).to_csv(log_file, mode='a', header=False, index=False)
        
        # Plot
        plt.clf()
        plt.plot(history_prices_live[-60:], label='Recent Price', color='blue')
        plt.scatter(len(history_prices_live), next_day_price, color='green', label='Next Day')
        plt.scatter(len(history_prices_live)+4, next_week_price, color='orange', label='Next Week')
        plt.scatter(len(history_prices_live)+20, next_month_price, color='red', label='Next Month')
        plt.title(f"FederalBK Live Price + Forecast ({now.strftime('%Y-%m-%d %H:%M:%S')})")
        plt.xlabel("Time Index")
        plt.ylabel("Close Price")
        plt.legend()
        plt.pause(0.1)
    
    time.sleep(60)
```
